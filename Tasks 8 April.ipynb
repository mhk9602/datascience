```python
#Task 1

import pandas as pd

# Define the file path
file_path = '/content/Global Superstore Orders 2016.xlsx'

# Load the dataset
df = pd.read_excel(file_path)

# Display the number of rows and columns
print("Shape of the dataset (rows, columns):")
print(df.shape)

# List all column names
print("\nColumn names:")
print(df.columns.tolist())

# Show the first five rows
print("\nFirst five rows of the dataset:")
display(df.head())
```

```python
#Task 2

# Display summary statistics for numerical columns, including Sales and Profit
print("\nSummary Statistics:")
display(df[['Sales', 'Profit']].describe().round(2))
```

```python
#Task 3

# Filter and display rows where Sales are greater than 500
sales_gt_500 = df[df['Sales'] > 500]
print("\nRows where Sales > 500:")
display(sales_gt_500.head())
print(f"Number of rows where Sales > 500: {sales_gt_500.shape[0]}")
print('\n')

# Change the filter value to 1000 and compare the number of rows
sales_gt_1000 = df[df['Sales'] > 1000]
print("\nRows where Sales > 1000:")
display(sales_gt_1000.head())
print(f"Number of rows where Sales > 1000: {sales_gt_1000.shape[0]}")
```

```python
# Task 4

# Sort orders by Profit from highest to lowest
df_sorted_by_profit = df.sort_values(by='Profit', ascending=False)

# Display the top five rows after sorting
print("\nTop five orders by Profit (highest to lowest):")
display(df_sorted_by_profit.head())
```

```python
#Task 5

# Group by Category and calculate total Profit
profit_by_category = df.groupby('Category')['Profit'].sum()
print("Total Profit by Category:")
display(profit_by_category)


# Group by Market and calculate total Profit
profit_by_market = df.groupby('Market')['Profit'].sum()
print("\nTotal Profit by Market:")
display(profit_by_market)
```

```python
#Task 6

# Add a new column 'Profit_Per_Sale'
df['Profit_Per_Sale'] = df['Profit'] / df['Sales']
print("\nFirst five rows of the DataFrame with 'Profit_Per_Sale' column:")
display(df.head(25))
```

```python
#Task 7

# Save the updated DataFrame to a new Excel file
df.to_excel("Cleaned_Global_Superstore.xlsx", index=False)
print("Updated dataset saved to 'Cleaned_Global_Superstore.xlsx'")
```

```python
import numpy as np

A = np.array([100, 120, 130, 150, 170, 160, 140, 110])
sorteddata = sorted (A)
lower_bound = np.percentile(A, 25)
upper_bound = np.percentile(A, 75)

middle_50_percent = A[(A >= lower_bound) & (A <= upper_bound)]

print("Original Array A:", A)
print("Sorted Data:", sorteddata)
print("25th Percentile:", lower_bound)
print("75th Percentile:", upper_bound)
print("Middle 50% of the array:", middle_50_percent)
```

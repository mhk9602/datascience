```python
import numpy as np
temps_c = np.array([18.5, 21.0, 24.3, 27.8, 31.5, 36.0, 40.2])

# Increase Celsius temperatures by 1.5
temps_c_increased = temps_c + 1.5

# Convert Celsius to Fahrenheit
temps_f = (temps_c * 9/5) + 32

# Cap Fahrenheit temperatures above 100 to 100
temps_f[temps_f > 100] = 100

print(f"Original temperatures in Celsius: {temps_c}\n")
print(f"Temperatures after increasing by 1.5 in Celsius: {temps_c_increased}\n")
print(f"Original temperatures in Fahrenheit: {temps_f}")
```

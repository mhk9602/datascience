```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

np.random.seed(0)
x = np.linspace(-5, 5, 40)
y_true = 3 * x**2 - 2 * x + 5
noise = np.random.randn(len(x)) * 5
y_noisy = y_true + noise

plt.scatter(x, y_noisy)
plt.title("Noisy data")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
def model(x, a, b, c):
    return a * x**2 + b * x + c

params, cov = curve_fit(model, x, y_noisy)
a_fit, b_fit, c_fit = params

print("Fitted values:")
print("a =", a_fit)
print("b =", b_fit)
print("c =", c_fit)

x_smooth = np.linspace(-5, 5, 200)
y_fit = model(x_smooth, a_fit, b_fit, c_fit)

plt.scatter(x, y_noisy, label="data")
plt.plot(x_smooth, y_fit, label="fitted curve")
plt.title("Data and fitted curve")
plt.legend()
plt.show()
```

```python
x_turn = -b_fit / (2 * a_fit)
y_turn = model(x_turn, a_fit, b_fit, c_fit)

print("Turning point x:", x_turn)
print("Turning point y:", y_turn)

plt.plot(x_smooth, y_fit, label="fitted curve")
plt.scatter(x, y_noisy, alpha=0.5, label="data")
plt.scatter(x_turn, y_turn, color="red", s=80, label="min / max point")
plt.title("Fitted curve with turning point")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import numpy as np

titanic = sns.load_dataset("titanic").copy()

print("Missing values per column:")
print(titanic.isnull().sum())

titanic["age"] = titanic["age"].fillna(titanic["age"].median())
titanic["embarked"] = titanic["embarked"].fillna(titanic["embarked"].mode()[0])
titanic = titanic.drop_duplicates()

print("\nAfter cleaning, missing values:")
print(titanic.isnull().sum())
```

```python
groups = pd.DataFrame({
    "class": ["First", "Second", "Third"],
    "group_size": [325, 285, 706]
})

titanic_merged = pd.merge(titanic, groups, on="class", how="left")
print(titanic_merged[["class", "group_size"]].head(3))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

num_cols = ["age", "fare", "sibsp", "parch"]
X = titanic_merged[num_cols].dropna()

X_scaled = StandardScaler().fit_transform(X)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

print("Explained variance ratio by PC1 and PC2,", pca.explained_variance_ratio_)
print("First 5 PCA rows,\n", X_pca[:5])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
titanic_merged[["age_minmax", "fare_minmax"]] = scaler.fit_transform(
    titanic_merged[["age", "fare"]]
)

print(titanic_merged[["age", "age_minmax", "fare", "fare_minmax"]].head())
```

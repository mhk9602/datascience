```python
import pandas as pd

# Series showing daily temperature for 5 days with day names as index
temperatures = pd.Series([22, 25, 23, 27, 24], index=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'])

# Make the text bold
print('\033[1mWeek Days  Daily Temperatures:\033[0m')
print(temperatures)

# Highest temperature
highest_temp = temperatures.max()
print(f'\nHighest Temperature: {highest_temp} °C')

# Average temperature using mean
average_temp = temperatures.mean()
print(f'Average Temperature: {average_temp:.2f} °C')
```

```python
import pandas as pd

# Create the first DataFrame for student information
student_info_data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']

    }
student_info_df = pd.DataFrame(student_info_data)
print('\033[1mStudent Information DataFrame:\033[0m')
print('\n')
print(student_info_df)
print('\n')

# Create the second DataFrame for student marks
student_marks_data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Eve'],
    'Score': [90, 85, 92, 78]

    }
student_marks_df = pd.DataFrame(student_marks_data)
print('\033[1mStudent Marks DataFrame:\033[0m')
print('\n')
print(student_marks_df)
print('\n')

# Merge the two DataFrames on the 'Name' column
# Using an 'outer' merge to keep all names from both DataFrames
merged_df = pd.merge(student_info_df, student_marks_df, on='Name', how='outer')

print('\033[1mMerged DataFrame (by Name):\033[0m')
print('\n')
print(merged_df)
```

```python
import pandas as pd

data = {
    'name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'age': [25, None, 35, None],
    'email': ['a@email.com', None, 'c@email.com', 'd@email.com'],
    'signup' : ['2023-01-10', '2023-02-15', '2023-03-20', '2023-03-25']
}

df = pd.DataFrame(data)

print('DataFrame created from data dictionary:')
print(df)
print('\n')

print('Missing values per column:')
print(df.isnull().sum())

# Calculate the median age
median_age = df['age'].median()

# Fill missing ages with the median age
df['age'] = df['age'].fillna(median_age)
print('\n')

print(f'DataFrame after filling missing ages with median ({median_age}):')
print(df)
print('\n')

# Remove rows with missing email addresses
df_cleaned = df.dropna(subset=['email'])
print('DataFrame after removing rows with missing email addresses:')
print(df_cleaned)
print('\n')

print('Data types of df_cleaned:')
print(df_cleaned.dtypes)
print('\n')
```
